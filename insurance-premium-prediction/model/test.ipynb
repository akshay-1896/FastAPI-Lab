```python
import pickle
import pandas as pd
import sys
sys.path.append('..')
from schema.user_input import UserInput

# import the ml model
with open('model.pkl', 'rb') as file:
    model = pickle.load(file)

# extracted by MLflow
MODEL_VERSION = '1.0.0'

user_input = {
  "age": 12,
  "weight": 100,
  "height": 1.55,
  "income_lpa": 10,
  "smoker": "true",
  "city": "Mumbai",
  "occupation": "retired"
}

def predict_output(user_input):
    # Validate and compute fields using Pydantic model
    validated_input = UserInput(**user_input)
    input_df = pd.DataFrame([validated_input.model_dump()])
    output = model.predict(input_df)[0]
    probabilities = model.predict_proba(input_df)[0]

    return output, probabilities

predict_output(user_input)[1]
```
